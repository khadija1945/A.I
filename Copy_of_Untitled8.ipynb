```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
df = pd.read_csv('train.csv')

# Display first few rows of the dataset to understand its structure
print("First 5 rows of the dataset:")
print(df.head())

# Basic information about the dataset
print("\nDataset information:")
print(df.info())

# Statistical summary of the dataset
print("\nStatistical summary of the dataset:")
print(df.describe())
```

```python
# Check for missing values
print("\nMissing values in the dataset:")
print(df.isnull().sum())

# Unique values in each column
print("\nUnique values in each column:")
print(df.nunique())
```

```python
# Handle missing values
# Fill 'Age' with median and 'Embarked' with mode
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop the 'Cabin' column due to a high number of missing values
df.drop(columns=['Cabin'], inplace=True)

# Check for duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Drop duplicates if any
df = df.drop_duplicates()

# Confirm cleaned dataset shape
print("\nDataset shape after cleaning:", df.shape)

# Convert 'Survived' to categorical type if necessary
df['Survived'] = df['Survived'].astype('category')
```

```python
# Set the style for the plots
sns.set(style='whitegrid')

# Plotting the countplot for 'Survived'
plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=df, palette='pastel', hue='Survived', legend=False)
plt.title('Survival Count (0 = No, 1 = Yes)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# Analysis and insights:
# This plot shows the count of passengers who survived and did not survive. We can see that more passengers did not survive.

# Plotting the distribution of 'Age'
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Analysis and insights:
# The age distribution shows that most passengers were between 20 and 40 years old, with fewer older passengers. This information helps us understand the age demographics of the passengers.

# Plotting the count of passengers by 'Embarked'
plt.figure(figsize=(8, 5))
sns.countplot(x='Embarked', data=df, palette='pastel', hue='Embarked', legend=False)
plt.title('Count of Passengers by Embarked')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()

# Analysis and insights:
# Most passengers embarked from port 'S' (Southampton), followed by 'C' (Cherbourg) and 'Q' (Queenstown). This gives us an idea of the distribution of passengers based on their boarding locations.

# Visualizing the survival rate by 'Pclass'
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=df, palette='pastel', hue='Pclass', legend=False)
plt.title('Survival Rate by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()

# Analysis and insights:
# Passengers in the 1st class had a higher survival rate compared to those in the 2nd and 3rd classes. This suggests that socio-economic status played a significant role in survival, with higher-class passengers having better access to lifeboats.

# Visualizing the survival rate by 'Sex'
plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', data=df, palette='pastel', hue='Sex', legend=False)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()

# Analysis and insights:
# Females had a significantly higher survival rate compared to males, indicating that women were given priority during the evacuation.

# Additional visualizations for more insights

# Visualizing the survival rate by 'Embarked'
plt.figure(figsize=(8, 5))
sns.barplot(x='Embarked', y='Survived', data=df, palette='pastel', hue='Embarked', legend=False)
plt.title('Survival Rate by Embarked')
plt.xlabel('Embarked')
plt.ylabel('Survival Rate')
plt.show()

# Analysis and insights:
# Passengers who embarked from 'C' (Cherbourg) had a higher survival rate compared to those from 'S' (Southampton) and 'Q' (Queenstown). This could be due to various factors, including socio-economic status and lifeboat access.

# Visualizing the survival rate by 'Age'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=df, palette='pastel', hue='Survived', legend=False)
plt.title('Survival Rate by Age')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

# Analysis and insights:
# Younger passengers, especially children, had a slightly higher survival rate. This suggests that younger individuals were also given priority during the evacuation.

# Visualizing the relationship between 'Fare' and 'Survived'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Fare', data=df, palette='pastel', hue='Survived', legend=False)
plt.title('Survival Rate by Fare')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()

# Analysis and insights:
# Passengers who paid higher fares had a higher survival rate, indicating that those in higher economic classes had better chances of survival.
```
